# 1. 线性代码

## 代码示例

```python
import numpy as np

# 定义两个向量
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# 向量加法
add_result = a + b
# 点乘
dot_result = np.dot(a, b)

print("vector a + b:", add_result)
print("vector a ⋅ b:", dot_result)

# 矩阵和向量
W = np.array([[1, 2, 3],
              [4, 5, 6]])
x = np.array([1, 2, 3])

# 矩阵乘向量
y = W.dot(x)
print("W x =", y)
```

# 2.微积分

## 代码示例

```python
import numpy as np

def f(w):
    return (w - 2)**2

def grad_f(w):
    return 2*(w - 2)

# 超参数
learning_rate = 0.1
num_iterations = 10

w = 10.0  # 初始化一个参数
for i in range(num_iterations):
    w = w - learning_rate * grad_f(w)
    print(f"第{i+1}次迭代: w = {w:.4f}, f(w) = {f(w):.4f}")
```

# 4.优化器

## 代码示例

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 准备数据
np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # 100个随机样本
true_w = 3.5
true_b = 1.2
y = true_w * X + true_b + np.random.randn(100, 1)  # 加一点噪声

# 2. 初始化参数
w = 0.0
b = 0.0
learning_rate = 0.1
epochs = 20

# 3. 训练
for epoch in range(epochs):
    # 计算模型输出
    y_pred = w * X + b
    
    # 计算损失 (均方误差)
    loss = np.mean((y_pred - y) ** 2)
    
    # 计算梯度
    dw = np.mean(2 * (y_pred - y) * X)
    db = np.mean(2 * (y_pred - y))
    
    # 更新参数
    w -= learning_rate * dw
    b -= learning_rate * db
    
    if (epoch+1) % 5 == 0:
        print(f"Epoch {epoch+1}: w = {w:.4f}, b = {b:.4f}, loss = {loss:.4f}")

# 4. 结果可视化
plt.scatter(X, y, label='Data')
plt.plot(X, w*X + b, color='red', label='Fitted line')
plt.legend()
plt.show()
```
